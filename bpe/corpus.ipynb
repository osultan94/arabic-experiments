```python
import re
```

```python
def preprocess_text_for_tokenization(text):
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    text = text.lower()
    return text
```

```python
titles['processed_text'] = titles['title'].apply(preprocess_text_for_tokenization)
```

```python
titles.head()
```

```python
len(titles_lst)
```

```python
prep_titles = titles['processed_text'].tolist()
```

```python
# save titles to text file where each line is a job title
with open('titles.txt', 'w') as f:
    for item in prep_titles:
        f.write("%s\n" % item)
```

```python
# read titles from text file to a list of job titles
with open('titles.txt', 'r') as f:
    titles_lst = f.read().splitlines()
```

```python
from tqdm.notebook import tqdm
```

```python
# write a function to check the number of unique tokens in all job titles
def check_unique_tokens(titles_lst):
    unique_tokens = set()
    for title in tqdm(titles_lst, desc="Checking unique tokens"):
        for token in title.split():
            unique_tokens.add(token)
    return len(unique_tokens)
```

```python
check_unique_tokens(prep_titles)
```
