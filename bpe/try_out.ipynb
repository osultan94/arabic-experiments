```python
import os
```

```python
# read all files in a directory, all in a list of strings
files_data = []
for root, dirs, files in os.walk('/home/omar/Documents/workspace/Byte-Pair Encoding/data/Khaleej-2004'):
    for file in files:
        with open(os.path.join(root, file), 'r') as f:
            files_data.append(f.read())
```

```python
len(files_data)
```

```python
# function to preprocess text, reduce multiple spaces into sinle space and trim spaces at the beginning and end
def preprocess_text(text):
    return ' '.join(text.split())
```

```python
new_data = [preprocess_text(file) for file in files_data]
```

```python
len(new_data)
```

```python
bpe = BPE()
# corpus = ["This is the first sentence.", "This is the second sentence.", "And this is the third sentence."]
# corpus = [sent.lower() for sent in corpus]
bpe.train(new_data, vocab_size=1000000, max_merges=10000)
```
